```python
# coding: utf-8

# ### Notebook to perform Zernike polynominal analysis of SKA1-LOW and LOFAR calibration

# In[1]:

%matplotlib notebook

import pylab

import matplotlib.pyplot as plt

pylab.rcParams['figure.figsize'] = (8.0, 8.0)
# pylab.rcParams['image.cmap'] = 'rainbow'

from teliono import *

import subprocess
print("Working from Git repository %s" % subprocess.check_output(["git", "describe", "--long", "--all"]))

def timestamp():
    from datetime import datetime
    print("%s %s" % (datetime.now().date(), datetime.now().time()))

```

```python
plt.clf()
ti=TelIono()
tsky=numpy.arange(1.0,12.0,0.1)
for DR in [1e4, 1e5, 1e6]:
    J=ti.J(tsky=tsky*30*24*3600,DR=DR)
    plt.semilogy(tsky,J,label='DR = %.1f dB' % (10.0*numpy.log10(DR)))
plt.xlabel('Observing time (months)')
plt.ylabel('Number of Zernikes required')
plt.title('Number of Zernikes required for given observing time')
plt.legend()
plt.show()
plt.savefig('Jneeded.pdf')
```

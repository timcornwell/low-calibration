```python
%matplotlib inline
import pylab
import matplotlib.pyplot as plt
pylab.rcParams['figure.figsize'] = (8.0, 8.0)
#pylab.rcParams['image.cmap'] = 'rainbow'
plt.switch_backend('pdf')

from telutil import *
from sources import *
from telinputs import *

import subprocess
print("Working from Git repository %s" % subprocess.check_output(["git",  "describe", "--long", "--all"]))


random.seed(781490893)

nstations=512
nnoll=1000

stationdiameter=35.0
rcore=0.5
rmin=0.0
hiono=300.0


print ("Freq, HWZ, tiono, imagenoise, visnoise, weight, piercethreshold, nsources")

freqs=numpy.zeros(len(numpy.arange(50.0,210.0,1.0)))
nsource=numpy.zeros(len(numpy.arange(50.0,210.0,1.0)))

plt.clf()
for bandwidth in [10**5, 10**5.5, 10**6.0, 10**6.5, 10**7]:
    i=0
    for mfreq in numpy.arange(50.0,210.0,1.0):
        freq=mfreq*1e6
        freqs[i]=freq
    
        wavelength=300.0/mfreq
        HWZ=hiono*wavelength/stationdiameter # Half width to Zero
        FOV=(0.5*wavelength/stationdiameter)**2

        rhalo=HWZ
        rmax=HWZ
    
        tiono=10.0 * numpy.power(wavelength / 3.0, -5.0 / 6.0)

        imgnoise, visnoise, weight, gainnoise = calculatenoise(tiono, freq, bandwidth, nstations,
                                                                     stationdiameter,
                                                                     doprint=False)
        piercethreshold=5.0*gainnoise
        nsources=FOV*sources().numbers(piercethreshold, freq)        

        imgthreshold=5.0*imgnoise
        nsources=int(FOV*sources().numbers(imgthreshold, freq))        



# Weight is the gain for a 1.0 Jy source"
        weight=1.0/gainnoise**2

        print ("%.1f, %.2f, %.2f, %.4f, %.3f, %.1f, %.3f, %.1f" % 
               (mfreq, HWZ, tiono, imgnoise, visnoise, weight, piercethreshold, nsources))
        nsource[i]=nsources
        i=i+1
    
    plt.plot(freqs/1e6, nsource, '-', label='BW %.1f MHz' %(bandwidth/1e6))
plt.legend()
plt.xlabel('Freq (MHz)')
plt.ylabel('Number of sources')
plt.title('Number of sources used in DFV')
plt.show()
plt.savefig('sourcesvsfreqDFV.pdf')
```
